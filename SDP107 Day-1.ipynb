```python
# Day 1 Assignment. Check Age field for null values and clean the data
#Data Cleaning,checking missing, find percentage, drop those
x=titanic['Age'].isnull() # Retrieve info about which all
l=[]# Initialise empty list to append index values of NULL values.

print("Total records:",len(x))
for i in range(len(x)):
    if x[i]:
        l.append(i) # Append Index value which has 'Age' values as NULL

print("Number of null age records:",len(l))
percentage=len(l)/len(x)
print("Percentage of null record:{0:.1f}%".format(percentage*100))

# Now remove the indexes whose values are null
titanic.drop(l,0)
# 0 indicates rows, 1 indicates delete columns. 
# Basically 0 deletes tuple, 1 deletes only column.
print("Dropped Values which had 'Age' as NULL values.")
```
